```python
def dummy_fn(df,dummy_col_list):
    for x in dummy_col_list:                                             
        dummies = pd.get_dummies(df[x], prefix =x, dummy_na = False)   #creating dummy binary values and removing null values
        df = df.drop(x, 1)                                              #dropping those existing columns
        df = pd.concat([df, dummies],axis =1)       # concat new dummy columns and existing df after dropping exisiting columns 
    return df
```
